## ✳️ Варіант 2: Алгоритм Евкліда (ділення та віднімання)
Функція знаходить найбільший спільний дільник (НСД) двох натуральних чисел за допомогою алгоритму Евкліда через **ділення та віднімання**.
Вбудовано перевірку правильності введених даних.


```python
def task_2():
    try:
        # Ввід чисел
        a, b = map(int, input("Введіть два натуральні числа через пробіл: ").split())
        if a <= 0 or b <= 0:
            print("❌ Числа мають бути додатними!")
            return

        original_a, original_b = a, b

        # Евклідів алгоритм (віднімання)
        while a != b:
            if a > b:
                a -= b
            else:
                b -= a

        print(f"✅ Найбільший спільний дільник {original_a} і {original_b} – це {a}")
    except ValueError:
        print("❌ Введіть два цілих числа через пробіл!")
```

```python
task_2()
```
